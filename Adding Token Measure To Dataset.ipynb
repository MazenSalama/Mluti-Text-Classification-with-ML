```python
# Method 2 Using Counter
```

```python
# Load NLP pkgs
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
```

```python
from collections import Counter
```

```python
# Function to Tokenize
def custom_counter(data):
    mytokens = word_tokenize(data)
    word_count_dict = Counter(w for w in mytokens if not w.lower() in stop_words)
    return word_count_dict.most_common()
```

```python
ex = "He is running with him, He got him"
```

```python
custom_counter(ex)
```

```python
df['Token_Measure'] = df['Section'].apply(custom_counter)
```
